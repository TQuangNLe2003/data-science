29. **TotalWorkingYears (Tổng số năm làm việc)**: Tổng số năm mà nhân viên đã làm việc.

30. **TrainingTimesLastYear (Số lần đào tạo trong năm)**: Số lần đào tạo mà nhân viên đã tham gia trong năm qua.

33. **YearsInCurrentRole (Số năm ở vị trí hiện tại)**: Số năm mà nhân viên đã ở trong vị trí công việc hiện tại.

Tuổi trung bình của nữ là 37.33 và nam là 36.65 và cả hai phân phối đều tương tự

```python
# Filter and select data for each gender
female_age = df[df['Gender'] == 'Female']['Age']
male_age = df[df['Gender'] == 'Male']['Age']

# Plotting the distribution of Age by gender
plt.figure(figsize=(10, 6))

# Female age distribution
sns.kdeplot(female_age, shade=True, color='pink', label='Female')
plt.axvline(x=female_age.mean(), linestyle='--', color='red', label='Mean Female Age')

# Male age distribution
sns.kdeplot(male_age, shade=True, color='lightblue', label='Male')
plt.axvline(x=male_age.mean(), linestyle='--', color='blue', label='Mean Male Age')

plt.title('Age Distribution by Gender')
plt.xlabel('Age')
plt.ylabel('Density')
plt.legend()
plt.show()
```

Biểu đồ phân phối độ tuổi theo giới tính. Mỗi phân phối sẽ được mô tả bằng một đường cong (KDE) với màu sắc khác nhau cho nam và nữ. Đường thẳng đứng đứng kẻ đứt ở giữa mỗi phân phối biểu thị giá trị trung bình của độ tuổi tương ứng với từng giới tính.

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Age', hue='Gender', kde=True)
plt.title('Age Distribution by Gender')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

### Analysis Generation and Education

#### Generation  
Biến 'Age' sẽ được dùng để xác định thế hệ



```python
# Xác định thế hệ dựa trên độ tuổi
def define_generation(age):
    if age <= 24:
        return 'Gen Z'
    elif age <= 40:
        return 'Millennials'
    elif age <= 56:
        return 'Gen X'
    else:
        return 'Baby Boomers'

df['Generation'] = df['Age'].apply(define_generation)
```

```python
plt.figure(figsize=(8, 4))
sns.countplot(x='Generation', hue='Attrition', data=df, palette='Set2')
plt.title('Attrition by Generation', fontsize=16)
plt.xlabel('Generation', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(title='Attrition', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

---> Biểu đồ Countplot đếm số lượng nhân viên nghỉ việc và không nghỉ việc trong mỗi thế hệ
- Thế hệ Millennials chiếm số lượng lớn nhất trong tổng số nhân viên, cũng như số lượng nhân viên nghỉ việc.
- Gen X và Baby Boomers có số lượng nhân viên nghỉ việc ít hơn so với Millennials, mặc dù số lượng tổng cộng của Gen X cũng khá cao.
- Gen Z, mặc dù có số lượng nhân viên ít nhất, nhưng có tỷ lệ nghỉ việc tương đối cao so với tổng số lượng

```python
# Tính tỉ lệ nhân viên nghỉ việc trong mỗi thế hệ
attrition_by_generation = df.groupby('Generation')['Attrition'].value_counts(normalize=True).unstack()
print("Tỉ lệ nhân viên nghỉ việc theo từng thế hệ:")
print(attrition_by_generation)

# Tính trung bình tuổi của những người nghỉ việc trong mỗi thế hệ
avg_age_by_generation = df[df['Attrition'] == 'Yes'].groupby('Generation')['Age'].mean()
print("\nTrung bình tuổi của nhân viên nghỉ việc trong mỗi thế hệ:")
print(avg_age_by_generation)
```

```python
# Tính trung bình số công ty từng làm của mỗi thế hệ
avg_companies_by_generation = df.groupby('Generation')['NumCompaniesWorked'].mean().reset_index()

# Vẽ biểu đồ cột
plt.figure(figsize=(10, 6))
sns.barplot(x='Generation', y='NumCompaniesWorked', data=avg_companies_by_generation, palette='viridis')
plt.title('Average Number of Companies worked for')
plt.xlabel('Generation')
plt.ylabel('Average Number of Companies worked for')
plt.xticks(rotation=45)
plt.show()
```

```python
# Nhóm dữ liệu theo bộ phận và tính toán thu nhập trung bình của mỗi nhóm
income_by_department = df.groupby('Department')['MonthlyIncome'].mean().reset_index()
```

```python
# Tính toán thu nhập trung bình của từng bộ phận và so sánh giữa những nhân viên đã rời bỏ và chưa rời bỏ
income_by_department_attrition = df.groupby(['Department', 'Attrition'])['MonthlyIncome'].mean().reset_index()

# Vẽ biểu đồ cột với thông số cụ thể trên mỗi cột
plt.figure(figsize=(12, 7))
barplot = sns.barplot(x='Department', y='MonthlyIncome', hue='Attrition', data=income_by_department_attrition)

# Chú thích từng thanh với giá trị cụ thể
for p in barplot.patches:
    barplot.annotate(format(p.get_height(), '.2f'),
                     (p.get_x() + p.get_width() / 2., p.get_height()),
                     ha = 'center', va = 'center',
                     xytext = (0, 9),
                     textcoords = 'offset points')

plt.title('Average Monthly Income by Department and Attrition Status')
plt.xlabel('Department')
plt.ylabel('Average Monthly Income')
plt.xticks(rotation=45)
plt.legend(title='Attrition')
plt.tight_layout()  # Điều chỉnh layout
plt.show()
```

---> Biểu đồ thể hiện thu nhập trung bình hàng tháng theo từng phòng ban và tình trạng nghỉ việc (Attrition) trong một tổ chức

```python
# Tạo các nhóm dựa trên mức độ hài lòng trong công việc
job_satisfaction_groups = df.groupby('JobSatisfaction')

# Tính toán thu nhập trung bình của mỗi nhóm
average_income_by_job_satisfaction = job_satisfaction_groups['MonthlyIncome'].mean().reset_index()

# Thiết lập giao diện biểu đồ
plt.figure(figsize=(10, 6))
sns.boxplot(x='JobSatisfaction', y='MonthlyIncome', data=df, palette='Set2')

# Thêm title và các nhãn
plt.title('Monthly Income by Job Satisfaction Level', fontsize=16)
plt.xlabel('Job Satisfaction Level', fontsize=14)
plt.ylabel('Monthly Income', fontsize=14)


# Hiển thị biểu đồ
plt.tight_layout()
plt.show()
```

---> Từ biểu đồ, có thể rút ra rằng mức độ hài lòng cao nhất (mức 4) dường như có liên quan đến mức thu nhập hàng tháng cao hơn. Tuy nhiên, do sự chồng chéo lớn giữa các phạm vi thu nhập và sự tồn tại của giá trị ngoại lệ, không thể khẳng định một cách chắc chắn mối liên hệ này chỉ qua biểu đồ hộp.

```python

# Tạo hai tập dữ liệu riêng cho nhân viên đã rời bỏ và nhân viên không rời bỏ
left_df = df[df['Attrition'] == 'Yes']
stayed_df = df[df['Attrition'] == 'No']

# Tính toán thu nhập trung bình của mỗi nhóm
left_mean_income = left_df['MonthlyIncome'].mean()
stayed_mean_income = stayed_df['MonthlyIncome'].mean()
# Tính toán thu nhập trung bình của nhân viên có xếp hạng hiệu suất cao và thấp
performance_mean_income = df.groupby('PerformanceRating')['MonthlyIncome'].mean()

# Tạo lưới 1x2 cho hai biểu đồ
plt.figure(figsize=(16, 6))

# Biểu đồ 1: So sánh thu nhập giữa nhân viên đã rời bỏ và nhân viên không rời bỏ tổ chức
plt.subplot(1, 2, 1)
sns.barplot(x=['Left Organization', 'Stayed in Organization'], y=[left_mean_income, stayed_mean_income], palette="Blues_d")
plt.title('Average Monthly Income by Attrition Status')
plt.ylabel('Average Monthly Income')
plt.xlabel('Attrition Status')

# Biểu đồ 2: So sánh thu nhập giữa nhân viên có xếp hạng hiệu suất cao và thấp
plt.subplot(1, 2, 2)
sns.barplot(x=performance_mean_income.index, y=performance_mean_income.values, palette="Set2")
plt.title('Average Monthly Income by Performance Rating')
plt.ylabel('Average Monthly Income')
plt.xlabel('Performance Rating')

plt.tight_layout()
plt.show()
```

---> Có thể có một mối quan hệ giữa thu nhập và tình trạng nghỉ việc, với những nhân viên có thu nhập cao hơn có khả năng ở lại tổ chức lớn hơn.

-  Những nhân viên có hiệu suất cao và/hoặc ở lại tổ chức có xu hướng có mức thu nhập cao hơn
